```python
# %load code/test6.py
import matplotlib.pyplot as plt
import matplotlib as mlb
import numpy as np
import toycode1 as toy
import seaborn as sns


'''============================================================================'''
def massDis(ax,x,y,m):
    PolyCollec = ax.hexbin(x,y)

    l = PolyCollec.get_offsets()[1,1]-PolyCollec.get_offsets()[0,1]

    HexArea = np.sqrt(3)/2*l**2
    m=m/HexArea
    ax.cla()

    pcm=ax.hexbin(x,y,C=m,cmap='inferno',reduce_C_function=np.sum)

    return pcm
    
'''=========================================================================='''

'''==========================================================================='''
def velDis(ax,x,y,v_z):
    cmap2 = plt.cm.get_cmap('seismic')
    pcm=ax.hexbin(x,y,C=v_z,cmap=cmap2)

    return pcm

'''============================================================================'''
sns.set(color_codes=True)

basePath = '/srv/cosmdatb/erebos/peng/sims.TNG/TNG300-1/output'
subhaloID = 0
snapNum = 99
#Type = ['spheroid','disc']
#string = Type[1]
#parData = toy.SDseparateCC(basePath,snapNum=135,subhaloID=subhaloID,a=1,H_z_0=1)['%s' %string]

parData = toy.cirparameter(basePath,snapNum=snapNum,ID=subhaloID,a=1,H_z_0=1)
''' make grid '''

c = parData['Coordinates']
m = parData['Masses']
v = parData['Velocities']
#cirPar = parData['cirPar']

''' get B/T'''

Nr = 2
Nc = 3
indexX = [0,0,1]
indexY = [1,2,2]
indexZ = [2,1,0]
stringX = ['x','x','y']
stringY = ['y','z','z']

fig,axs = plt.subplots(Nr,Nc,sharex=True,sharey=True)
#fig.suptitle('0.1 ID: %d (%s)' %(subhaloID,string))
fig.suptitle('ID: %d' %(subhaloID))
for i in np.arange(Nr):
    pcm = [0,0,0]
    for j in np.arange(Nc):
        ax = axs[i,j]
        if i == 0:
            x = c[:,indexX[j]]
            y = c[:,indexY[j]]
            pcm[j] = massDis(ax,x,y,m)
        else:
            x   = c[:,indexX[j]]
            y   = c[:,indexY[j]]
            v_z = v[:,indexZ[j]]
            pcm[j] = velDis(ax,x,y,v_z)
            
        ax.set_xlabel(stringX[j]+'$\ (kpc/h)$')
        ax.set_ylabel(stringY[j]+'$\ (kpc/h)$')

    vmin = min(image.get_array().min() for image in pcm)
    vmax = max(image.get_array().max() for image in pcm)
    if i == 0:
        norm = mlb.colors.LogNorm(vmin=vmin, vmax=vmax)
    else:
        if abs(vmin) < vmax:
            vmin = -1*vmax
        else:
            vmax = abs(vmin)
        norm = mlb.colors.Normalize(vmin=vmin, vmax=vmax)
    for im in pcm:
        im.set_norm(norm)
    cb = fig.colorbar(pcm[0],ax=[axs[i,:]])
    if i == 0:
        cb.set_label('$\log\Sigma\ (M_{\odot} h/kpc^2)$')
    else:
        cb.set_label('$V\ (km/s)$')
                
plt.show()
```

```python
def HiH2RadialProfile(ID,basePath,snapNum,a=1,h=0.677):
    Nr = 1
    Nc = 2
    proj  = ['_2d']
    col   = ['_neutral_H','_mol_GK11']
    sx = ['$HI+H_{2}$','$H_{2}\ $GK11']
    subhaloInf = il.groupcat.loadSingle(basePath,snapNum,subhaloID=ID)
    R_starHalf = (subhaloInf['SubhaloHalfmassRadType'][4] * a) / h
    R_gasHalf  = (subhaloInf['SubhaloHalfmassRadType'][0] * a) / h 
    #haloId  = subhaloInf['SubhaloGrNr']
    #haloInf = il.groupcat.loadSingle(basePath,snapNum,haloID=haloId)
    #R_200c  = (haloInf['Group_R_Crit200'] * a) / h
    
    Data = {}
    with h5py.File('/srv/cosmdatb/erebos/peng/sims.TNG/TNG300-1/Supp/hih2_galaxy_099.hdf5','r') as f:
        if ID not in f['id_subhalo']:
            raise Exception('this subhalo is not included')
            
        indexR = np.where(np.array(f['id_subhalo']) == ID)[0][0]
        binsD   = np.array(f['profile_bins'][indexR,:]) - R_gasHalf 
        indexC_lim = np.min(np.where(binsD >= 0)) + 1 
        
        for i in np.arange(Nr):
            for j in np.arange(Nc):
                Data['profile_f'+col[j]+proj[i]] = np.array(f['profile_f'+col[j]+proj[i]][indexR,:indexC_lim])
                
            Data['profile_gas_rho'+proj[i]] = np.array(f['profile_gas_rho'+proj[i]][indexR,:indexC_lim])
                
        Data['profile_bins'] = np.array(f['profile_bins'][indexR,:indexC_lim])
        
    '''plot'''
    
    
    
    #fig = plt.figure()
    #gs = fig.add_gridspec(Nr, Nc, hspace=0, wspace=0)
    #axs = gs.subplots(sharex=True, sharey=True)
    fig, axs = plt.subplots(Nr, Nc,sharex=True, sharey=True)
    
    for i in np.arange(Nr):
        for j in np.arange(Nc):
            x = Data['profile_bins']
            if j == 0:
                y = np.log10(Data['profile_f'+col[j]+proj[i]] * Data['profile_gas_rho'+proj[i]] + 1)
            else:
                y = np.log10(Data['profile_f'+col[0]+proj[i]] * Data['profile_gas_rho'+proj[i]] * Data['profile_f'+col[j]+proj[i]] + 1)
            
            axs[j].plot(x,y,label=proj[i].strip('_'))
            axs[j].set_title(sx[j],{'fontsize':'small'})
            axs[j].set(xlabel='R/kpc', ylabel='$\log_{10}(\Sigma / (M_{\odot}/kpc^2))$')
            axs[j].vlines([R_starHalf,R_gasHalf], 0, 1, transform=axs[j].get_xaxis_transform(), 
                            colors='r',linestyles=['dotted','dashed'])
            axs[j].legend(fontsize='small')
            
    for i in np.arange(Nr):
        for j in np.arange(Nc):
            axs[j].label_outer()
    fig.suptitle('ID:%d' %ID)
    plt.savefig('/srv/cosmdatb/erebos/yingzhong/fig/HiH2Fig%d'%ID,dpi=250)
    #plt.show()
```

```python
def transform(sSFR,M_star):
    if sSFR == -20:
        return 0
    else:
        return 10**sSFR * M_star

def loadData(arr_id,arr_sSFR,arr_Mstar,s,a=1,h=0.677):
    
    Nr = 1
    Nc = 1
    proj  = ['_2d']
    Data = {}
    M    = {} 
    
    for i in np.arange(Nc):
        Data['R_stellar']      = np.zeros(0)
        Data['M_HI'+proj[i]]   = np.zeros(0)
        Data['M_H2'+proj[i]]   = np.zeros(0)
        Data['gas'+proj[i]]    = np.zeros(0)
        Data['SFR']            = np.zeros(0)
        Data['M_star']         = np.zeros(0)
    
    with h5py.File('/srv/cosmdatb/erebos/peng/sims.TNG/TNG300-1/Supp/hih2_galaxy_099.hdf5','r') as f:
        for ID in arr_id:
            if len(np.where(np.array(f['id_subhalo']) == ID)[0]) == 0:
                continue
            
            subhaloInf = il.groupcat.loadSingle(basePath,snapNum,subhaloID=ID)
            R_starHalf = (subhaloInf['SubhaloHalfmassRadType'][4] * a) / h
            R_gasHalf  = (subhaloInf['SubhaloHalfmassRadType'][0] * a) / h
            
            indexR     = np.where(np.array(f['id_subhalo']) == ID)[0][0]
            indexF     = np.where(arr_id == ID)[0][0]
            
            M_star     = arr_Mstar[indexF]
            SFR        = transform(arr_sSFR[indexF],M_star)
            binsL      = np.array(f['profile_bins'][indexR,:]) - 0*R_starHalf
            binsR      = np.array(f['profile_bins'][indexR,:]) - 3*R_starHalf
                
            #index_limL = np.min(np.where(binsL >= 0))
            if len(np.where(binsL >= 0)[0])==0:
                index_limL = 49
            else:
                index_limL = np.min(np.where(binsL >= 0))
            
            if len(np.where(binsR >= 0)[0])==0:
                index_limR = 50
            else:
                index_limR = np.min(np.where(binsR >= 0)) + 1
                
            #index_limR = 51
        
            for j in np.arange(Nc):
                M['gas'+proj[j]]  = np.array(f['profile_gas_rho'+proj[j]][indexR,index_limL:index_limR]) *\
                np.array(f['profile_bins_area'][indexR,index_limL:index_limR])
                
                M['M_H2'+proj[j]] = np.array(f['profile_f_mol_GK11'+proj[j]][indexR,index_limL:index_limR])*\
                np.array(f['profile_f_neutral_H'+proj[j]][indexR,index_limL:index_limR])*\
                M['gas'+proj[j]]
                
                M['M_HI'+proj[j]] = (1 - np.array(f['profile_f_mol_GK11'+proj[j]][indexR,index_limL:index_limR]))*\
                np.array(f['profile_f_neutral_H'+proj[j]][indexR,index_limL:index_limR])*\
                M['gas'+proj[j]]
                
                M['R_stellar']    = R_starHalf
                M['SFR']          = SFR
                M['M_star']       = M_star
                
            for i in np.arange(Nc):
                Data['gas'+proj[i]]  = np.append(Data['gas'+proj[i]],np.sum(M['gas'+proj[i]]))
                Data['M_HI'+proj[i]] = np.append(Data['M_HI'+proj[i]],np.sum(M['M_HI'+proj[i]]))
                Data['M_H2'+proj[i]] = np.append(Data['M_H2'+proj[i]],np.sum(M['M_H2'+proj[i]]))
                Data['R_stellar']    = np.append(Data['R_stellar'],M['R_stellar'])
                Data['SFR']          = np.append(Data['SFR'],M['SFR'])
                Data['M_star']       = np.append(Data['M_star'],M['M_star'])
    
    np.savez('/srv/cosmdatb/erebos/yingzhong/selData/data/HiH2'+s+'outer_GK11_0_3.npz',
             gas=Data['gas'+proj[0]],M_HI=Data['M_HI'+proj[0]],
             M_H2=Data['M_H2'+proj[0]],R_stellar=Data['R_stellar'],SFR=Data['SFR'],
            M_star=Data['M_star']) 
    
    return
    
def QuenchMechDig(basePath,snapNum,a=1,h=0.677):
    
    N=2
    fMmass = np.load('/srv/cosmdatb/erebos/yingzhong/selData/data/haveBH.npz')
    f1 = np.load('/srv/cosmdatb/erebos/yingzhong/selData/data/haveBH_025.npz')
    Str= ['form','quench'] 
    index  = reduce(np.intersect1d,
                    (np.where(np.array(fMmass['sSFR']) > -11),
                     np.where(np.array(fMmass['ratioS_T']) < 0.25)))
    #index1 = np.where(np.array(f1['M_star']) >= 10**11)[0]
    
    arr_Id   = [np.array(fMmass['subHaloID'][index]),np.array(f1['subHaloID'])]
    arr_sSFR = [np.array(fMmass['sSFR'][index]),np.array(f1['sSFR'])]
    arr_Mstar= [np.array(fMmass['M_star'][index]),np.array(f1['M_star'])]
    
    #DataAll = [0,0]
    for i in np.arange(N):
        loadData(arr_Id[i],arr_sSFR[i],arr_Mstar[i],s='%s' %(Str[i]))
    
    
```

```python
%matplotlib inline
import numpy as np
import illustris_python1 as il
from matplotlib import pyplot as plt
import seaborn as sns
'''plot'''
N = 2
Label   = ['$sSFR > 10^{-11}\ yr^{-1}$ & $S/T<0.25$','$sSFR < 10^{-11}\ yr^{-1}$ & $S/T<0.25$']
DataAll = [0,0]
median  = [{},{}]

DataAll[0] = dict(np.load('/srv/cosmdatb/erebos/yingzhong/selData/data/HiH2formouter_GK11_0_50.npz'))
DataAll[1] = dict(np.load('/srv/cosmdatb/erebos/yingzhong/selData/data/HiH2quenchouter_GK11_0_50.npz'))
c = ['#1f77b4','#ff7f0e']
ls= ['--','-.']
Type  = ['starform','quench']
keys  = ['gas','M_HI','M_H2']
xName = ['gas','HI','H_{2}']
IndexFakeForm   = np.where(DataAll[0]['M_H2']<10**6)[0]
IndexFakeQuench = np.where(DataAll[1]['M_H2']<10**6)[0]

for key in keys:
    DataAll[0][key] = np.delete(DataAll[0][key],IndexFakeForm)
    median[0][key]  = np.median(DataAll[0][key])

for key in keys:
    DataAll[1][key] = np.delete(DataAll[1][key],IndexFakeQuench)
    median[1][key]  = np.median(DataAll[1][key])

        
p = 0
fig,axs = plt.subplots(1, 1,sharex=True, sharey=True)  
gname = keys[p]
for i in np.arange(N):
    z      = np.log10(DataAll[i][gname])
    medlog = np.log10(median[i][gname])
    axs = sns.distplot(z,label=Label[i])
    axs.axvline(medlog,color='k',ls=ls[i],label='$\log_{10}(M_{%s}/M_{\odot})=%3.1f $' %(Type[i],medlog))
    #axs.axvline(medlog,color=c[i],ls=ls[i],label='%3.1f' %(medlog))



axs.set(xlabel='$log_{10}(M_{%s}/M_{\odot})$' %xName[p])
axs.legend(fontsize='x-small')
#plt.savefig('/srv/cosmdatb/erebos/yingzhong/fig2/gas_sSFR_0_50_1c',dpi=250)
plt.show()
```
